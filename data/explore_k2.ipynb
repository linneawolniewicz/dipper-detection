```python
from astropy.io import ascii,fits
import numpy as np
import matplotlib.pyplot as plt
import os
```

# Load one fits file and plot the lightcurve

```python
dir = "k2/"
filename = dir + "k2_65.fits"
df = fits.open(filename)
data = df[1].data
t = np.array(data['TIME'])
f = np.array(data['PDCSAP_FLUX'])
```

```python
plt.plot(t,f,'k')
plt.show()

# print the columns and size of data and df
print(data.columns)
print(data.size)
print(df.info())
```

# Plot every lightcurve in the k2 directory

```python
dir = "../data/k2/"

for i in range(78):
    filename = dir + "k2_" + str(i) + ".fits"

    assert os.path.exists(filename), f"{filename} does not exist"
    
    df = fits.open(filename)
    data = df[1].data
    t = np.array(data['TIME'])
    f = np.array(data['PDCSAP_FLUX'])

    plt.title(f"k2_{i}, # time steps: {len(t)}")
    plt.plot(t,f,'k')
    plt.show()
```
